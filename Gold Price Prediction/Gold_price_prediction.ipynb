```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
df=pd.read_csv('gld_price_data.csv')
df.head(5)
```

```python
df.tail(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
#Calculating correlation
df1=df.drop('Date', axis=1)
```

```python
correlation=df1.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8},cmap='Greens')
```

```python
print(correlation['GLD'])
```

```python
#distribution of gold
sns.distplot(df1['GLD'],color='Green')
```

```python
#Splitting the target from the features
X=df1.drop('GLD',axis=1)
Y=df1['GLD']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Split into test and train
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
# Model Training
regressor=RandomForestRegressor(n_estimators=100)
```

```python
regressor.fit(X_train,Y_train)
```

```python
#evaluation
data_prediction=regressor.predict(X_test)
```

```python
#R^2 error
error_score=metrics.r2_score(Y_test, data_prediction)
print(error_score)
```

```python
#comparision between actual and prediction
Y_test=list(Y_test)
```

```python
plt.plot(Y_test, color='green', label='Actual Values')
plt.plot(data_prediction, color='red', label='Predictions')
plt.title("actua; Vs Predictions")
plt.xlabel("Number of Values")
plt.ylabel("GLD price")
plt.legend()
plt.show()
```
