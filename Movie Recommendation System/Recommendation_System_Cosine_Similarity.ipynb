```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('dataset.csv')
movies.head(10)
```

```python
movies.columns
```

```python
movies.shape
```

```python
movies['tags']=movies['genre']+movies['overview']
```

```python
movies.head(3)
```

```python
new_movies=movies[['id','title','genre','overview','tags']]
```

```python
new_movies=new_movies.drop(columns=['genre','overview'])
```

```python
new_movies.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000,stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_movies['tags'].values.astype('U')).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cs=cosine_similarity(vector)
```

```python
cs
```

```python
new_movies[new_movies['title']=='The Shawshank Redemption']
```

```python
dist=sorted(list(enumerate(cs[0])),reverse=True, key=lambda cs:cs[1])
```

```python
dist
```

```python
for i in dist[0:5]:
  print(new_movies.iloc[i[0]].title)
```

```python
def recommend(movies):
  index=new_movies[new_movies['title']==movies].index[0]
  distance=sorted(list(enumerate(cs[index])),reverse=True, key=lambda cs:cs[1])
  for i in distance[0:5]:
    print(new_movies.iloc[i[0]].title)
```

```python
recommend('Iron Man')
```
